### 1. Compute daily returns for the stock for the entire sample duration ###

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# get data and clean it
df = pd.read_csv('https://raw.githubusercontent.com/kpace1111/portfoliomanagement/main/TSLA.csv')
df = df[['Date', 'Adj Close']]
df.rename(columns={'Date' : 'date', 'Adj Close' : 'price_t'}, inplace=True)
df['price_t-1'] = df['price_t'].shift(1)

# Calculate Returns
df['returns'] = df['price_t'].pct_change(1)
df.set_index('date', inplace=True)
df.head()
```

### 2. Plot the daily price of the stock and its daily returns in a line chart / graph ###

```python
# Plotting Daily Stock Price
df['price_t'].plot(label = 'Price', 
                         figsize = (12, 8), 
                         c = 'b', 
                         grid = True, 
                         title = 'Tesla Stock Price')
plt.legend();

# Saving Plot
plt.savefig('Tesla_Stock_Price')
```

```python
# Plotting Daily Stock Price
df['returns'].plot(label = 'Return', 
                         figsize = (12, 8), 
                         c = 'b', 
                         grid = True, 
                         title = 'Tesla Stock Returns',
                         alpha = 0.7)
plt.legend();

# Saving Plot
plt.savefig('Tesla_Stock_Return')
```
