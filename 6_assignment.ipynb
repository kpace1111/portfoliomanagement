1. Extract daily price data for 10 stocks of your choice, for a 5 year timeframe.

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
#import data
df = pd.read_csv('https://raw.githubusercontent.com/kpace1111/portfoliomanagement/main/10stocks_assignment.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')] #remove unamed columns
df['date'] = pd.to_datetime(df['date']) # create dates as timestamp
df.set_index('date', inplace=True)
#df.head()
```

2. Compute the daily returns of each stock. Remember to check and ensure that your prices are stacked at the same dates!

```python
# calculate returns of all 10 stocks
returns_df = df.pct_change(1)
returns_df
```

### Questions for this assignment:
## 1. Calculate the Variance Covariance Matrix for your 10 stocks, and create a vector of equal weights for the 10 assets.



```python
# Create vector of even weights (omega) 
num_stocks = 10
weights = [1/num_stocks] * num_stocks
#weights

#calculate VCV Matrix - sigma
vcv_matrix = returns_df.cov() #works out var and cov for all 10 stocks
vcv_matrix
```

## 2. Estimate the total risk of your equally weighted 10 asset portfolio.

```python
# take transposed of vector of weights and multiply by product of the VCV matrix and the vector of weights
var_p = np.dot(np.transpose(weights), np.dot(vcv_matrix, weights))
sd_p = np.sqrt(var_p)
sd_p #daily
sd_p_annual = sd_p * np.sqrt(250)
sd_p_annual
```

```python
#Compare these risks to the individual risks of the stocks (annually)
individual_risks = np.std(returns_df) * np.sqrt(250)
individual_risks
```
